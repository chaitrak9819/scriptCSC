```python
# -*- coding: utf-8 -*-
"""
Created on Tue Nov  9 11:57:17 2021

@author: thosh
"""


def row_unused_digit(grid, valid_digit, i):

    for k in range(9):
        if grid[i][k] > 0:
            valid_digit[grid[i][k] - 1] = 0
            

def column_unused_digit(grid, valid_digit, j):

    for k in range(9):
        if grid[k][j] > 0:
            valid_digit[grid[k][j] - 1] = 0
            
def findTheBlock(columnNumber):
    
        if(columnNumber<=2):
            
            # first
            return 0
            
        elif (columnNumber<=5):
            
            # second
            return 3
        else:
            
            return 6
            # third
    
            
def block_unused_digit(grid, valid_digit, i, j):
    
    if(i <=2):
     
        colCt = findTheBlock(j)
        rowct = 0
            
    elif(i <= 5):
        
        colCt = findTheBlock(j)
        rowct = 3
        
    else:
        
        colCt = findTheBlock(j)
        rowct = 6
        
    
    for val1 in range(rowct,rowct+3):
        
        for val2 in range(colCt,colCt+3):
            
            if grid[val1][val2] > 0:
                
                valid_digit[grid[val1][val2] - 1] = 0
                

def sudoku2str(gridToPrint):
    
      print('---------------------------------------------------')
      
      
      spaceString = ' '
      
      for row in range(len(gridToPrint)):
          
          stringToConcatenate = ''
          
          if row == 3 :
              
              print('---------------------------------------------------')
              
          elif row == 6:
              
              print('---------------------------------------------------')
              
          for col in range(len(gridToPrint[row])):
              
              if col ==0 :
                  
                  if gridToPrint[row][col] == 0:
                      
                      stringToConcatenate = stringToConcatenate + '|'  + spaceString * 3  + '_'
                      
                  else:
                      
                      stringToConcatenate = stringToConcatenate + '|'  + spaceString * 3  + str(gridToPrint[row][col])
                      
              elif  col ==3 or col == 6:
                  
                    if gridToPrint[row][col] == 0:
                      
                      stringToConcatenate = stringToConcatenate + spaceString * 3 + '|'  + spaceString * 3  + '_'
                      
                    else:
                         
                        stringToConcatenate = stringToConcatenate + spaceString * 3 + '|'  + spaceString * 3  + str(gridToPrint[row][col])
                 
                    
              elif col == 8:
                  
                  if gridToPrint[row][col] == 0:
                     
                      stringToConcatenate = stringToConcatenate + spaceString * 3 + '_' +  spaceString * 3 + '|'
                      
                  else:
                      
                      stringToConcatenate = stringToConcatenate + spaceString * 3 + str(gridToPrint[row][col]) +  spaceString * 3 + '|'
                    
              else:
                  
                  
                  if gridToPrint[row][col] == 0:
                     
                      stringToConcatenate = stringToConcatenate + spaceString * 3 + '_'
                      
                  else:
                      
                      stringToConcatenate = stringToConcatenate + spaceString * 3 + str(gridToPrint[row][col])
                  
          print(stringToConcatenate)
          
      print('---------------------------------------------------')
                

        
def best_cell(listOfValidDigitsAndIndex):
    
    storeVal = listOfValidDigitsAndIndex[0][2].count(1)
    
    retlist = [listOfValidDigitsAndIndex[0][0],listOfValidDigitsAndIndex[0][1],listOfValidDigitsAndIndex[0][2]]
    
    for val1 in listOfValidDigitsAndIndex:
        
        if (storeVal > val1[2].count(1)):
            
            storeVal = val1[2].count(1)
            
            retlist = [val1[0],val1[1],val1[2]]

            
    return retlist

def possible_number(grid,i,j):
    
    
    valid_digit_For_PossibleSolution = [1,1,1,1,1,1,1,1,1]
                    
    row_unused_digit(grid,valid_digit_For_PossibleSolution,i)
                
    column_unused_digit(grid,valid_digit_For_PossibleSolution,j)
                    
    block_unused_digit(grid,valid_digit_For_PossibleSolution,i,j)
    
    list_of_possibleValues = []
    
    for vals in range(len(valid_digit_For_PossibleSolution)):
        
        if valid_digit_For_PossibleSolution[vals] ==1:
            
            list_of_possibleValues.append(vals+1)
            
    return list_of_possibleValues


def solve_sudoku(grid):
    
    listToStore = []
    
    for row in range(len(grid)):
    
        for col in range(len(grid[row])):
            
            if grid[row][col] ==0:
                
                    valid_digit = [1,1,1,1,1,1,1,1,1]
                    
                    row_unused_digit(grid,valid_digit,row)
                
                    column_unused_digit(grid,valid_digit,col)
                    
                    block_unused_digit(grid,valid_digit,row,col)
                    
                    listToStore.append([row,col,valid_digit])
                    
                
    
    if (len(listToStore) != 0):
        
        retVal = best_cell(listToStore)
        
        if 1 not in retVal[2]:
            
            return 0
    
        for vls in range(len(retVal[2])):
            
            if retVal[2][vls] == 1:
               
                grid[retVal[0]][retVal[1]] = vls + 1
              
                retvalzeroOrOne = solve_sudoku(grid) 
                    
                if retvalzeroOrOne ==0:
                        
                    grid[retVal[0]][retVal[1]] = 0
                        
                else:
                        
                    return grid
                    
        return 0

    return grid
        
```
